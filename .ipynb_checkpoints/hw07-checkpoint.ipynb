### Выбросы

```python
df.describe()
```

```python
df['age_filled_mean'].hist(bins = 100)
```

```python
# Природа датасета не предполагает наличие выбросов. 
# Если только не считать выбросом то, чем мы заполненили пропуски.
# Но раз уж надо, посчитаем выбросы по возрасту пассажиров.
```

```python
q1 = df['age_filled_mean'].quantile(0.25)
q3 = df['age_filled_mean'].quantile(0.75)
iqr = q3-q1
lower_dound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print(q1, q3)
print(lower_dound, upper_bound)
```

```python
outliers = df[(df['age_filled_mean'] > upper_bound)]
```

```python
outliers
```

```python
df[(df['age_filled_mean'] <= upper_bound)]
```

### Нормализация

```python
df['norm_age'] = (df['age_filled_mean'] - df['age_filled_mean'].min()) / (df['age_filled_mean'].max() - df['age_filled_mean'].min())
```
